### Word Cloud 1 - Simple

We will create a Word Cloud of top words from Wonder Woman Movie. We will use the movie script provided in [this website](https://www.springfieldspringfield.co.uk/movie_script.php?movie=wonder-woman-2017). We will need to remove Stop Words from the script before creating the cloud. wordcloud library provides a list of stop words. We will use that list.

```python
from wordcloud import WordCloud, STOPWORDS
import matplotlib
import matplotlib.pyplot as plt
from PIL import Image
import numpy as np

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (16.0, 9.0)
```

```python
# Reading the script
script = open("wonderwoman.txt", "r").read()
# Set of Stop words
stopwords = set(STOPWORDS)
stopwords.add("will")
# Create WordCloud Object
wc = WordCloud(background_color="white", stopwords=stopwords, 
               width=1600, height=900, colormap=matplotlib.cm.inferno)
# Generate WordCloud
wc.generate(script)
# Show the WordCloud
plt.figure()
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```

### Word Cloud 2 - With Mask

We can also create Word Clouds with custom masks.

```python
# Reading the script
script = open("wonderwoman.txt", "r").read()
# Set of Stop words
stopwords = set(STOPWORDS)
stopwords.add("will")
# Add mask
wonderwoman_mask = np.array(Image.open("wonderwoman.jpg"))

# Create WordCloud Object
wc = WordCloud(background_color="white", stopwords=stopwords, mask=wonderwoman_mask,
               width=1600, height=900, colormap=matplotlib.cm.inferno)
# Generate WordCloud
wc.generate(script)
# Show the WordCloud
plt.figure()
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```

It is clear that "Diana" is the most repeated word.

Similarly for "The Dark Knight(2008)" movie with a Batman symbol mask. [Script Link](https://www.springfieldspringfield.co.uk/movie_script.php?movie=dark-knight-the-batman-the-dark-knight)

```python
script = open("darknight.txt", "r").read()
# Set of Stop words
stopwords = set(STOPWORDS)
stopwords.add("will")
# Add mask
batman_mask = np.array(Image.open("batman.png"))
from matplotlib.colors import LinearSegmentedColormap
colors = ["#484848", "#000000", "#0060A8", "#FFF200", "#303030"]
cmap = LinearSegmentedColormap.from_list("mycmap", colors)
# Create WordCloud Object
wc = WordCloud(background_color="white", stopwords=stopwords, mask=batman_mask,
               colormap=cmap)
# Generate WordCloud
wc.generate(script)
# Show the WordCloud
plt.figure()
plt.imshow(wc, interpolation="bilinear")
plt.axis("off")
```

### Word Cloud 3 - Colored Mask

We will create Word Cloud of "Captain America: Civil War" [script](https://www.springfieldspringfield.co.uk/movie_script.php?movie=captain-america-civil-war) with following mask.

This method colorizes the cloud with average color in the area.

```python
from wordcloud import ImageColorGenerator

script = open("civilwar.txt", "r").read()
# Set of Stop words
stopwords = set(STOPWORDS)
stopwords.add("will")
# Add mask
civilwar_mask = np.array(Image.open("civilwar.jpg"))

# Create WordCloud Object
wc = WordCloud(background_color="white", stopwords=stopwords, mask=civilwar_mask)
# Generate WordCloud
wc.generate(script)
# Image color generator
image_colors = ImageColorGenerator(civilwar_mask)
# Show the WordCloud
plt.figure()
plt.imshow(wc.recolor(color_func=image_colors), interpolation="bilinear")
plt.axis("off")
```
